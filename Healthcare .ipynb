## HealthCare Dataset

This is a Health care dataset with features such as - 
    
- name : Name of the patient
- age : Age of the patient
- gender : Gender of the patient
- blood_type : Blood type of the patient
- medical_condition : Name of the condition or disease
- date_of_admission : Date of joining in the hospital
- doctor : Name of the doctor
- hospital : Name of the Hospital
- insurance_provider : Name of the insurance provider
- billing_amount : Amount charged for the patient
- room_number : Allocated room number of the patient
- admission_type : Name of the Admission
- discharge_date : Date on which the patient is discharged
- medication : Name of the medication used for treatment
- test_results : Test result of the patient upon treatment

#### Observations of Univariate Analysis

- The count plots for **Blood Type**, **Admission Type**, and **Insurance Provider** show that these variables are distributed fairly evenly, with no single category dominating significantly.

### Bivariate Analysis

- Compare variables Age and Admission Type

```python
# Box plot for Age by Admission Type

plt.figure(figsize=(10, 6))
sns.boxplot(x='admission_type', y='age', data=data, hue='admission_type', palette='coolwarm', dodge=False)

plt.title("Age Distribution by Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Age")
plt.show()
```

- Compare variables Admission Type and Insurance Provider

```python
# Heatmap for Admission Type and Insurance Provider
admission_insurance = data.groupby(['admission_type', 'insurance_provider']).size().unstack().fillna(0)

# Plot
plt.figure(figsize=(12, 6))
sns.heatmap(admission_insurance, annot=True, fmt=".0f", cmap='YlGnBu', cbar=True)

plt.title("Admission Type by Insurance Provider")
plt.xlabel("Insurance Provider")
plt.ylabel("Admission Type")
plt.show()
```

#### Observations for Bivariate Analysis

- The distribution of Age across different Admission Types reveals that the interquartile range (25th to 75th percentile) lies between ages 35 and 70, indicating that most patients fall within this age group for various admission types.
- The data shows a strong association between the Cigna insurance provider and the Urgent admission type, indicating that a significant proportion of urgent admissions are covered by Cigna.
- In contrast, there is a weaker association between the Aetna insurance provider and the Urgent admission type, suggesting that fewer urgent admissions are associated with Aetna.

### Multivariate Analysis

```python
# Correlation Heatmap
numerical_cols = ['age', 'billing_amount']  # Include other numerical columns if applicable
correlation = data[numerical_cols].corr()

plt.figure(figsize=(8, 5))
sns.heatmap(correlation, annot=True, cmap='Blues', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Grouped Bar Chart
grouped_data = data.groupby(['gender', 'admission_type'])['billing_amount'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='admission_type', y='billing_amount', hue='gender', data=grouped_data, palette='viridis')
plt.title("Average Billing Amount by Gender and Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Average Billing Amount")
plt.legend(title="Gender")
plt.show()
```

#### Observations for Multivariate Analysis
- here appears to be no significant correlation between Age and Billing Amount, suggesting that billing amounts are not influenced by the patient's age.
- The distributions of Admission Types across male and female patients are generally similar, with only slight variations observed between genders. However, these differences are minimal and may not be statistically significant.

### Questions

1. Billing Amount Across Insurance Providers and Admission Types

```python
# Billing Amount by Insurance Provider and Admission Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='insurance_provider', y='billing_amount', hue='admission_type', data=data, palette='Set2')

plt.title("Billing Amount by Insurance Provider and Admission Type")
plt.xlabel("Insurance Provider")
plt.ylabel("Billing Amount")
plt.xticks(rotation=45)
plt.show()
```

- The Billing Amount shows slight variations based on the Admission Type for a given Insurance Provider, but these changes are not significant.

2. Blood Type and Medical Conditions

```python
# Blood Type vs Medical Condition
blood_condition = data.groupby(['blood_type', 'medical_condition']).size().unstack().fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(blood_condition, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title("Blood Type vs Medical Conditions")
plt.xlabel("Medical Condition")
plt.ylabel("Blood Type")
plt.show()
```

- There is only a very slight difference between the values for Medical Condition and Blood Type, indicating minimal variation in their distribution or relationship.

3. Length of Stay by Gender and Admission Type

```python
# Calculate Length of Stay
data['date_of_admission'] = pd.to_datetime(data['date_of_admission'])
data['discharge_date'] = pd.to_datetime(data['discharge_date'])
data['length_of_stay'] = (data['discharge_date'] - data['date_of_admission']).dt.days

plt.figure(figsize=(10, 6))
sns.barplot(x='admission_type', y='length_of_stay', hue='gender', data=data, palette='Spectral')
plt.title("Length of Stay by Gender and Admission Type")
plt.xlabel("Admission Type")
plt.ylabel("Average Length of Stay")
plt.legend(title="Gender")
plt.show()
```

- There is no notable difference in the Length of Stay based on Gender or Admission Type.

4. Correlation Between Age, Billing Amount, and Medications

```python
# Scatter Plot Matrix
numerical_cols = ['age', 'billing_amount']  # Include 'medication' if numeric
sns.pairplot(data[numerical_cols], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle("Scatter Plot Matrix", y=1.02)
plt.show()
```

- There is no apparent relationship between Age and Billing Amount, suggesting that age does not influence billing amounts.
